# Log Returns v.s. Arithmetic Returns  

First, why use logarithm of returns rather than price or raw returns? 
- Rather than using **prices**, the benefit of using returns is that it *normalizes* the dataset. When interpreting the eventual covariance matrix, it is easier to comprehend the variables when both are measured in percentages
- Rather than using raw **returns**, the idea behind using log returns is because stock prices are not normally distributed and have long right hand tails. However, after taking the log pf the returns, prices are conveniently normally distributed  

   
In addition, log returns prove to have additional benefits.
- When returns are very small, the log returns are close in value to the raw returns
- For compounding returns, different times are additive.  

Example:
- Suppose we have stock ABC Inc. that is USD 1.00 at T=0, USD 0.80 at T=1, and USD 1.00 at T=3
- The arithmatic mean of the returns day over day is a 2.5% increase
- The logarithmic mean, which is the sum of the logs, is a 0% return -> **Much more accurate for compounding returns**  

*Extra Note: Average log returns over a time period of length t only require 3 calculations which has a time complexity of $\mathbf{O}(1)$, where as average arithmatic returns require n calculations which has a time complexity of $\mathbf{O}(t)$*

*Source: https://quantivity.wordpress.com/2011/02/21/why-log-returns/*



```python
#Removing Stocks w/ Limited History & Changing Dataframe to Numpy Array
df_close = df_close.dropna(axis = 'columns') #Removal of stocks
df_change = df_close.apply(lambda x: np.log(x) - np.log(x.shift(1)))  #Log returns
df_change = df_change.dropna() #Dropping first row - cannot compute of T(0) without prior date stock price
```

## Portfolio of Minimum Variance  $ \\ $

The goal of this portfolio is to minimize variance (volatility) while maximizing our returns. By definition, we can do this by maximizing the Sharpe ratio, however, there are other ratios that we intend to explore as well. This is subject to the sum of the weights of each stock in the pf equaling 100%.

Portfolio Variance (Historial Volatility):  $\sigma_p^2 = \mathbf{w^{ T}}\sum\mathbf{w}$  
Constraint (Sum of Weights must equal 1): $\mathbf{w^{ T}}\mathbf{1} = 1$  

### Legrangian Multipliers
- Differentiating with respect to $\mathbf{w}$  
$
L = \mathbf{w^{ T}}\sum\mathbf{w} + \lambda(\mathbf{w^{ T}}\mathbf{1} - 1) \\
\frac{\partial \mathcal L}{\partial \mathbf w} = 2 \sum \mathbf{w} + \lambda\mathbf{1} = 0 \\
\mathbf{w} = -\frac{1}{2}\lambda\sum^{-1}\mathbf{1}
$
